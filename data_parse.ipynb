```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
start = 'https://tracklock.gg/matches/'
num = 5800000 # 6300004-6329649
games = []
for i in range(20000):
    url = start + str(num + i)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    etc = soup.findAll('p', class_='text-2xl font-bold text-white font-sapient')
    if not etc:
        continue
    heroes = []
    heroes.append(num + i)
    heroes.append(*etc[3])
    found_heroes = soup.find_all('span', class_='font-medium')
    for hero in found_heroes:
        heroes.append(hero.text)
    games.append(heroes)

len(games)
```

```python
df = pd.DataFrame(games)
A_team = ['hero_1_A', 'hero_2_A', 'hero_3_A', 'hero_4_A', 'hero_5_A', 'hero_6_A']
B_team = ['hero_1_B', 'hero_2_B', 'hero_3_B', 'hero_4_B', 'hero_5_B', 'hero_6_B']
df.columns = [['matchid', 'winner'] + A_team + B_team]
df.dropna(inplace=True)
df['matchid'] = df['matchid'].astype(int)
df
```

```python
df.to_csv('data/new_matches.csv', index=False)
```
