```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
start = 'https://tracklock.gg/matches/'
num = 6300004 # 6300004-6329649
games = []
for i in range(50000):
    url = start + str(num + i)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    etc = soup.findAll('p', class_='text-2xl font-bold text-white font-sapient')
    if not etc:
        continue
    heroes = []
    pack = []
    heroes.append(num + i)
    heroes.append(*etc[1])
    heroes.append(*etc[3])
    all = soup.findAll('td')
    for data in all:
        if data.text == 'The Amber Hand' or data.text == 'The Sapphire Flame':
            continue
        if len(heroes) == 15:
            games.append(heroes)
            break
        pack.append(data.text)
        if len(pack) == 8:
            heroes.append(pack)
            pack = []
len(games)
```

```python
len(games)
```

```python
df = pd.DataFrame(games)
df.columns = [str(x) for x in range(15)]
df
```

```python
for column in df.columns:
    if column in ('0', '1', '2'):
        continue
    df[[f'hero_{int(column) - 2}', f'nickname_{int(column) - 2}', f'lane_{int(column) - 2}', f'K/D/A_{int(column) - 2}', f'souls_{int(column) - 2}', f'SPM_{int(column) - 2}', f'LH/DH_{int(column) - 2}', f'level_{int(column) - 2}']] = df[column].tolist()
df.drop(df.columns[3:15], axis=1, inplace=True)
df = df.copy()
df
```

```python
for i in range(1, 13):
    df[[f'K_{i}', f'D_{i}', f'A_{i}']] = df[f'K/D/A_{i}'].str.split('/', expand=True)
    df[[ f'LH_{i}', f'DH_{i}']] = df[f'LH/DH_{i}'].str.split('/', expand=True)
df = df.copy()
df.rename({'0': 'matchid', '1': 'duration', '2': 'winner'}, axis=1, inplace=True)
df
```
